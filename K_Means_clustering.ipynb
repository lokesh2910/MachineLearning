```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Generating random data for clustering
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
# Visualizing the generated data
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Generated Data")
plt.show()
```

```python
# Fitting KMeans to the data (let's try 4 clusters, as we generated 4 centers)
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
# Getting the cluster centers
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='X')  # Cluster centers
plt.title("K-Means Clustering Result")
plt.show()
```

```python
# Evaluating using inertia (sum of squared distances of samples to their closest cluster center)
print(f'Inertia: {kmeans.inertia_}')
```
